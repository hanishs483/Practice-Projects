```python
page7=requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YF6yfa9KjIVs")
page7
```

```python
dayss=[]
for i in days:
    dayss.append(i.text.replace('\n',''))
dayss[0:7]
```

```python
short=soup7.find_all('p',class_="short-desc")
short[0:7]
```

```python
shortt=[]
for i in short:
    shortt.append(i.text.replace('\n',''))
shortt[0:7]
```

```python
temp=soup7.find_all('p',class_="temp temp-high")
temp[0:5]
```

```python
Descr=[]
for i in desc:
    Descr.append(i.text.replace('\n',''))
Descr[0:7]
```
